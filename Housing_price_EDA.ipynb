```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
# Basic Data Exploration
print(data.head())
print(data.info())
print(data.describe())
```

```python
# Distribution of Median Income
plt.figure(figsize=(10, 6))
sns.histplot(data['median_income'], bins=30, kde=True)
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot of Median Income vs Median House Value
plt.figure(figsize=(10, 6))
plt.scatter(data['median_income'], data['median_house_value'], alpha=0.5)
plt.title('Median Income vs Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
# Categorical Variable: population Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data['population'])
plt.title('population Distribution')
plt.xlabel('population')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot of Total Rooms
plt.figure(figsize=(8, 6))
sns.boxplot(data['total_rooms'])
plt.title('Box Plot of Total Rooms')
plt.xlabel('Total Rooms')
plt.show()
```

```python

# Comparison of Median House Value across population
plt.figure(figsize=(10, 6))
sns.boxplot(x='population', y='median_house_value', data=data)
plt.title('Median House Value across population')
plt.xlabel('population')
plt.ylabel('Median House Value')
plt.xticks(rotation=45)
plt.show()
```
